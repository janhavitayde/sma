DATA CLEANING

```python
import sqlite3
import pandas as pd
import re
```

```python
# Read data from Excel file
df = pd.read_excel('/content/Youtube scrapped data.xlsx')
```

```python
df.head()
```

```python
# Define a function to preprocess text
def preprocess_text(text):
    # Remove URLs
    text = re.sub(r'http\S+', '', text)
    # Remove non-alphanumeric characters and extra whitespaces
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Convert to lowercase
    text = text.lower()
    return text

# Apply preprocessing to relevant columns
text_columns = ['Title', 'Range']  # Add more columns if needed
for col in text_columns:
    df[col] = df[col].apply(preprocess_text)

# Display the first few rows of the cleaned and preprocessed DataFrame
print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

DATA CLEANING

```python
# Read data from Excel file
df = pd.read_excel('/content/Youtube scrapped data.xlsx')
```

```python
df.head()
```

```python
df.info()
```

```python
# Function to convert string with suffix to numeric value
def convert_to_numeric(value):
    if isinstance(value, str):
        if 'K' in value:
            return int(float(value[:-1]) * 1000)  # Convert '2.5K' to 2500
        elif 'M' in value:
            return int(float(value[:-1]) * 1000000)  # Convert '3.7M' to 3700000
        else:
            return int(value)
    else:
        return int(value)

# Apply the function to convert 'Range2' column
df['Range2'] = df['Range2'].apply(convert_to_numeric)
```

```python
# Now, you can plot the histogram for 'Range1'
plt.hist(df['Range2'], bins=5, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Range 2')
plt.ylabel('Frequency')
plt.title('Histogram of Range 2')

# Show plot
plt.show()
```

```python
 import seaborn as sns

# Plot density graph for 'Range2'
sns.kdeplot(df['Range2'].dropna(), color='skyblue')

# Add labels and title
plt.xlabel('Range 2')
plt.ylabel('Density')
plt.title('Density Plot of Range 2')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot scatter plot for 'Range2' against index
plt.scatter(df.index, df['Range2'], color='skyblue')

# Add labels and title
plt.xlabel('Index')
plt.ylabel('Range 2')
plt.title('Scatter Plot of Range 2')

# Show plot
plt.show()
```

```python
import seaborn as sns

# Plot box plot for 'Range2'
sns.boxplot(x=df['Range2'], color='skyblue')

# Add title
plt.title('Box Plot of Range 2')

# Show plot
plt.show()
```

```python
import seaborn as sns

# Plot violin plot for 'Range2'
sns.violinplot(x=df['Range2'], color='skyblue')

# Add title
plt.title('Violin Plot of Range 2')

# Show plot
plt.show()
```
